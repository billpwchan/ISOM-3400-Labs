### Problem 1: Duplicate App Names / Data Record

```python
dupCount = data.groupby("App").count()>1
dupCount
```

```python
dupCountFilter = dupCount[dupCount["Category"]]
dupCountFilter
```

```python
data[data["App"].isin(dupCountFilter.index)].sort_values(by="App")
```
