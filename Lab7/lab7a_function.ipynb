```python
# Cinema Seating plan
# Task 4
# Print function

# Write a function named printSeats() to print the seating plan of a house

seats = [
    ["01", "02", "03", "04", "05", "06", "07"],
    ["01", "02", "03", "xx", "xx", "06", "07"],
    ["xx", "xx", "03", "04", "05", "06", "07"],
    ["01", "02", "xx", "xx", "xx", "xx", "xx"]
]

# expectation:
'''
printSeats(seats,2,4,1)
printSeats(seats,1,5,1)
printSeats(seats,3,1,3)
printSeats(seats,3,4)
printSeats(seats)


[               Screen               ]

A [01][02]    [03][04][05][06]    [07]
B [01][02]    [03][xx][xx][06]    [07]
C [xx][xx]    [03][04][05][06]    [07]
D [01][02]    [xx][xx][xx][xx]    [xx]

[               Screen               ]

A [01]    [02][03][04][05][06]    [07]
B [01]    [02][03][xx][xx][06]    [07]
C [xx]    [xx][03][04][05][06]    [07]
D [01]    [02][xx][xx][xx][xx]    [xx]

[               Screen               ]

A [01][02][03]    [04]    [05][06][07]
B [01][02][03]    [xx]    [xx][06][07]
C [xx][xx][03]    [04]    [05][06][07]
D [01][02][xx]    [xx]    [xx][xx][xx]

[             Screen             ]

A [01][02][03]    [04][05][06][07]
B [01][02][03]    [xx][xx][06][07]
C [xx][xx][03]    [04][05][06][07]
D [01][02][xx]    [xx][xx][xx][xx]

[           Screen           ]

A [01][02][03][04][05][06][07]
B [01][02][03][xx][xx][06][07]
C [xx][xx][03][04][05][06][07]
D [01][02][xx][xx][xx][xx][xx]
'''




def printSeats(seatList, a = None, b = 0, c = 0):

    if a == None:
        a = len(seatList[len(seatList)-1])
        
    line_space = 4 * a
    if b > 0:
        line_space += 4*(b+1)
    if c > 0:
        line_space += 4*(c+1)
        
    screen_line = "[%s]"% "Screen".center(line_space)
    rows = len(seatList)
    cols = len(seatList[len(seatList)-1])
       
    print(screen_line)
    print()
    
    for row, i in zip(seatList, range(rows)):
        seatFormat = chr(i+65)+" "+"[%s]"*a
        if b > 0:
            seatFormat = seatFormat + "    " + "[%s]"*b
        if c > 0:
            seatFormat = seatFormat + "    " + "[%s]"*c
        print(seatFormat%tuple(row))
        
    print()


printSeats(seats,2,4,1)
printSeats(seats,1,5,1)
printSeats(seats,3,1,3)
printSeats(seats,3,4)
printSeats(seats)

seats1 = [
    ["01", "02", "03", "04", "05", "06", "07", "07"],
    ["01", "02", "03", "xx", "xx", "06", "07", "07"],
    ["xx", "xx", "03", "04", "05", "06", "07", "07"],
    ["01", "02", "xx", "xx", "xx", "xx", "xx", "07"],
    ["xx", "xx", "03", "04", "05", "06", "07", "07"],
    ["01", "02", "xx", "xx", "xx", "xx", "xx", "07"]
]


printSeats(seats1,2,4,2)
printSeats(seats1,1,5,2)
printSeats(seats1,3,1,4)
printSeats(seats1,3,4,1)
printSeats(seats1)
```
