````python
# Currency Example
import datetime
import random

class Currency:
    def __init__(self, currency, value):
        self.currency = currency
        self.history= []
        self.history.append((datetime.datetime(2018,8,(101//24+1), 101%24, 0, 0, 0) , value))
        #self.history.append((datetime.datetime.today() , value))

    
    def currentValue(self):
        self.history.sort()
        return self.history[-1][1]
    
            
    def addHistory(self, time, value):
        self.history.append((time , value))
        self.history.sort()
        
    def plotGraph(self):
        from matplotlib import pyplot as plt
        plt.plot(dict(self.history).keys(),dict(self.history).values())
        plt.title("Currency HKD 1 to " + self.currency)
        plt.xlabel("Time")
        plt.ylabel(self.currency)
        plt.show()
        
    def fromHKD(self, hkd):
        return hkd * self.currentValue()
    
    def toHKD(self, amount):
        return amount / self.currentValue()
````

````python
jpy = Currency("JPY", 14.07866)

# add some random history data
for x in range(100):
    jpy.addHistory(datetime.datetime(2018,8,(x//24+1), x%24, 0, 0, 0), round(random.uniform(14.00, 14.20),2))

jpy.plotGraph()

print(jpy.fromHKD(100.7)) # HKD 100.7 = JPY ???
print(jpy.toHKD(100)) # JPY 100 = HKD ???
````

````python
usd = Currency("USD", 0.12740)

# add some random history data
for x in range(100):
    usd.addHistory(datetime.datetime(2018,8,(x//24+1), x%24, 0, 0, 0), round(random.uniform(0.127, 0.128),5))

usd.plotGraph()

print(usd.fromHKD(200)) # HKD 200 = USD ???
print(usd.toHKD(100)) # USD 100 = HKD ???
````
