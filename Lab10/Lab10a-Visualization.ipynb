### Example 2: Count Plot using Seaborn

From the pokemon dataset, I would like to use a bar chart to the count of each primary type of (non-)legendary pokemons.

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(15, 10))
sns.countplot(
    x = "Type 1",
    hue="Legendary", # this option to split the bar into 2 -> Legendary / non-legendary
    data = pokemon,
    order = pokemon['Type 1'][pokemon['Legendary']==False].value_counts(ascending=True).index,
)

plt.ylabel("Count / Frequency", fontsize = 20)
plt.xlabel("Primary Type", fontsize = 20)
plt.title('Pokemon Primary Type Count Plot',color = 'blue',  fontsize = 20)
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

Try to use countplot with the above dataset "tips" to plot a bar chart.
- count the number of record in the dataset for each day (x) for smokers / non-smokers (hue)
- Title: Transactions per day

```python
# ans
plt.figure(figsize=(12, 6))
sns.countplot(
    x = "day",
    hue="smoker", # this option to split the bar into 2 -> Legendary / non-legendary
    data = tips,
)

plt.ylabel("Count", fontsize = 20)
plt.xlabel("Day", fontsize = 20)
plt.title('Transactions per day',color = 'blue',  fontsize = 20)
plt.show()
```

### Example 3: Visualizing linear relationships - lmplot

Many datasets contain multiple quantitative variables, and the goal of an analysis is often to relate those variables to each other.

```python
#basic syntax
sns.lmplot(x="total_bill", y="tip", data=tips);
```

### Conditioning on other variables

How does the relationship between these two variables change as `sex` change?


```python
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips);
```

Findings: Both Male and Female paid similar ratio of tip for a meal

=> `Sex` does not affect a lot

```python
# ans:
sns.lmplot(x="total_bill", y="tip", hue="time", data=tips);
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
sns.lmplot(x="total_bill", y="tip", hue="day", data=tips);
```
